LINEAR REGRESSION USING SYNTHETIC DATA

```python
import numpy as np
X = 2 * np.random.rand(100,1)
y= 4 + 3 * X + np.random.rand(100,1)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y,"b")
plt.axis([0,2,0,15])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()#This line is the most important
lin_reg.fit(X,y)
y_pred = lin_reg.predict(X)
```

```python
plt.scatter(X,y,color ="blue")
plt.plot(X,y_pred, color ="red",linewidth=3)
```

```python
X_new = np.array([[0],[1.5],[2]])
lin_reg.predict(X_new)
```

```python
plt.scatter(X,y,color="blue")
plt.plot(X,y_pred,color="red",linewidth=2)
plt.plot(X_new,lin_reg.predict(X_new),"y+",markersize=29)
plt.xticks()
plt.yticks()
plt.show()
#yellow stars show the y values predicted for three points
```
